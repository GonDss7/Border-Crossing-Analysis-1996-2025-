```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('/content/Border_Crossing_Entry_Data.csv')
display(df.head())
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
print(df.info())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
display(df.info())
```

```python
port_code_counts = df.groupby('Port Name')['Port Code'].nunique()
display(port_code_counts[port_code_counts > 1])
```

```python
duplicate_rows = df[df.duplicated()]
print("Number of duplicate rows:", duplicate_rows.shape[0])
```

```python
df.drop_duplicates(inplace=True)
print("Number of duplicate rows after removal:", df.duplicated().sum())
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
print("Number of missing values after removal:\n", df.isnull().sum())
```
